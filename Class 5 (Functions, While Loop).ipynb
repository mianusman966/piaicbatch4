# Replicate SUM Function

```python
def add(any_list):
    sum1 = 0
    for i in any_list:
        sum1 += i
    return sum1 #we write return keyword from below of variavle which we show or return
```

```python
add(lst) #my replicate function and its out put
```

```python
sum(lst) #Python builtin function and its output
```

```python
#Class Task make my own Average Function
def avg(any_list):
    avg = 0
    sum1 = 0
    count = 0
    for i in any_list:
        sum1 += i
        count += 1
        avg = sum1/count
    return avg
```

```python
avg(lst)
```

```python
list5 = [5,5,5,5,5]
```

```python
avg(list5)
```

# Replica of Length

```python
def length(a_list):
    count = 0
    for i in a_list:
        count += 1
    return count
```

```python
length(lst)
```

# Scope

```python
a = 3 # any variable define outside a fun = Global Variable, Permanent Variable that access anywere
def div(x):
    return x/a
```

```python
div(3)
```

```python
# any variable define inside a fun = Local Variable, not access anywere
def divi(x):
    d=5 #local variable
    return x/d
```

```python
print(d)
```
